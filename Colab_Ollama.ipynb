## Exemplo de requisição

```python
import requests
url = 'https://0de9e072a91f.ngrok-free.app' ## tem que verificar na conta ngrok a url
url = f"{url}/api/generate"

# Cabeçalhos da requisição
headers = {
    "Content-Type": "application/json"
}

# Corpo da requisição (payload)
payload = {
    "model": "llama3",  # Nome do modelo
    "prompt": "Qual foi minha ultima mensagem?",  # Mensagem a ser enviada
    "stream": False  # Desativa o streaming para receber a resposta completa de uma vez
}

# Fazer a requisição POST
response = requests.post(url, headers=headers, json=payload)

# Verificar se a requisição foi bem-sucedida
if response.status_code == 200:
    # Exibir a resposta da API
    print("Resposta da API:", response.json())
else:
    print("Erro na requisição:", response.status_code, response.text)
```
